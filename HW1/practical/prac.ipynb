Imports

```python
from random import *
import numpy as np
import matplotlib.pyplot as plt
```

Monte Carlo function

This works according to The function : <br>
4 * alpha = pi * 1

```python
def calculate_pi(itterations,random_func):
    points_in_circle = 0
    for i in range(itterations):
        x = random_func(-1,1) 
        y = random_func(-1,1) 
        if x*x + y*y <= 1 :
                points_in_circle +=1
                
    return 4 * points_in_circle/itterations
```

Find pi

```python
itterations = 10000000
random_func = np.random.uniform
print(calculate_pi(itterations,random_func))
```

```python
def draw_graph(x_values, y_values):
    plt.plot(x_values, y_values)
    plt.xlabel('itteration')
    plt.ylabel('pi estimate')  
    # Show the plot
    plt.show()
```

```python
def get_and_draw(itterations,random_func,estimate_func):
    all_itterations = []
    pi_estimation = []
    for itteration in range(1,itterations):
        all_itterations.append(itteration)
        pi_estimation.append(estimate_func(itteration,random_func))
    draw_graph(all_itterations,pi_estimation)
```

```python
itterations = 1000
random_func = np.random.uniform
get_and_draw(itterations=itterations,random_func=random_func,estimate_func=calculate_pi)
```

In this simulation we used the <b>uniform distribution</b> to find our random numbers

```python
itterations = 1000
random_func = np.random.normal
get_and_draw(itterations=itterations,random_func=random_func,estimate_func=calculate_pi)
```

Monte Carlo - Estimate Ellipse Area

```python
def calculate_ellipse_area(itterations,random_func):
    points_in_ellipse = 0
    for i in range(itterations):
        x = random_func(-4,4) 
        y = random_func(-4,4) 
        if 5*x*x + 5*y*y - 6*x * y - 32 <= 0 :
                points_in_ellipse +=1
                
    return 64 * points_in_ellipse/itterations
```

```python
itterations = 10000000
random_func = np.random.uniform
print(calculate_ellipse_area(itterations,random_func))
```

```python
itterations = 3000
random_func = np.random.uniform
get_and_draw(itterations=itterations,random_func=random_func,estimate_func=calculate_ellipse_area)
```
