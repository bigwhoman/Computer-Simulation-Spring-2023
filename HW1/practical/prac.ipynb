# Monte Carlo function

This works according to The function : <br>
4 * alpha = pi * 1

```python
pi_estimates = []
pi_itterations = []
```

```python
def calculate_pi(random_func):
    points_in_circle = 0
    itterations = 0
    estimate = 0 
    while True:
        itterations += 1 
        x = random_func(-1,1) 
        y = random_func(-1,1) 
        if x*x + y*y <= 1 :
                points_in_circle +=1
        estimate = 4 * points_in_circle / itterations
        pi_estimates.append(estimate)
        pi_itterations.append(itterations)
        if abs(estimate - 3.1415) <= 0.000001 : 
            break 
                
    return estimate,itterations 
```

# P1 - Find pi

```python
# itterations = 10000000
random_func = np.random.uniform
pi = calculate_pi(random_func)
print(f'found pi = {pi[0]:.5f} in {pi[1]} steps !!')
```

```python
def draw_graph(x_values, y_values):
    plt.plot(x_values, y_values)
    plt.xlabel('itteration')
    plt.ylabel('pi estimate')  
    # Show the plot
    plt.show()
```

```python
def get_and_draw(itterations,random_func,estimate_func):
    pi = estimate_func(itterations,random_func)
    draw_graph(pi[0],pi[1])
```

# P2 - Draw Convergence Graph

```python
draw_graph(pi_itterations,pi_estimates)
```

In this simulation we used the <b>uniform distribution</b> to generate our random numbers

# P3 - Find pi with itterations

```python
def find_pi_with_itters(itterations,random_func):
    points_in_circle = 0
    itter_list = []
    estimation_list = []
    for i in range(1,itterations+1):
        x = random_func(-1,1) 
        y = random_func(-1,1) 
        if x*x + y*y <= 1 :
                points_in_circle +=1
        itter_list.append(i)
        estimation_list.append(4 * points_in_circle/i)
                
    return itter_list,estimation_list
```

```python
p2 = find_pi_with_itters(pi[1],random_func = np.random.uniform)
print(p2[1][pi[1]-1])
draw_graph(p2[0],p2[1])
```

```python
p3 = find_pi_with_itters(pi[1],random_func = np.random.uniform)
print(p3[1][pi[1]-1])
draw_graph(p3[0],p3[1])
```

In the above graphs, we could see that both our estimates, converge to the same number but the starting point may differ because of the random generated points. <br>
Also we could conclude that there is a significant importance on generating the <br> right random numbers </br>

# P4 - Using Normal Distribution

```python
itterations = 1000
random_func = np.random.normal
get_and_draw(itterations=itterations,random_func=random_func,estimate_func=find_pi_with_itters)
```

# P5 - Monte Carlo Estimate Ellipse Area

In this part, we should find the erea of Ellipse.<br>
We do this by calculating a rectangle around the ellipse and then estimating the erea of our ellipse.<br>
The generated points are in a 8 * 8 rectangle that is centered on (0,0), points are generated in range (-4,4) for height and length so that the rectangle can <b>fully cover</b> our ellips.

```python
def calculate_ellipse_area(itterations,random_func):
    points_in_ellipse = 0
    for i in range(itterations):
        x = random_func(-4,4) 
        y = random_func(-4,4) 
        if 5*x*x + 5*y*y - 6*x * y - 32 <= 0 :
                points_in_ellipse +=1
                
    return 64 * points_in_ellipse/itterations
```

```python
itterations = 10000000
random_func = np.random.uniform
print(calculate_ellipse_area(itterations,random_func))
```

```python
itterations = 3000
random_func = np.random.uniform
get_and_draw(itterations=itterations,random_func=random_func,estimate_func=calculate_ellipse_area)
```

As you could see, by itterating further, our answers would converge to a single number and thus the itteration number is acceptable.
